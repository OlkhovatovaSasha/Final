```python
#среднее часов в месяц и повышения за последние 5 лет
promotion_yes = hr[hr['promotion_last_5years']==1]['average_montly_hours']
promotion_no = hr[hr['promotion_last_5years']==0]['average_montly_hours']
print(sts.ttest_ind(promotion_no,promotion_yes))
```

t-value равно 0.4340 (не превышает 1,95) и p-value 0.66425 (не меньше чем 0,05), 

значит выборки "средние часы в месяц" и "повышения за последние 5 лет" статистически похожи.

```python
#среднее часов в месяц и номер проекта
promotion_yes1 = hr[hr['promotion_last_5years']==1]['number_project']
promotion_no1 = hr[hr['promotion_last_5years']==0]['number_project']
print(sts.ttest_ind(promotion_no1,promotion_yes1))
```

t-value равно 0.7426 (не превышает 1,95) и p-value 0.457 (не меньше чем 0,05), 

значит выборки "средние часы в месяц" и "номер проекта" статистически похожи.

```python
#несчастный случай на работе и последней оценки
Work_accident_yes = hr[hr['Work_accident']==1]['last_evaluation']
Work_accident_no = hr[hr['Work_accident']==0]['last_evaluation']
print(sts.ttest_ind(Work_accident_no,Work_accident_yes))
```
